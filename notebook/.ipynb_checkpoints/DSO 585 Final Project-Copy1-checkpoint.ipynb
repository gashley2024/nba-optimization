```python
combined_df_team1 = team1_df.merge(
    team2_df,
    left_on='opponent_team_uid_team1',
    right_on='team_uid_team2',
    suffixes=('_team1', '_team2')
)

combined_df_team2 = team2_df.merge(
    team1_df,
    left_on='opponent_team_uid_team2',
    right_on='team_uid_team1',
    suffixes=('_team2', '_team1')
)

combined_df = pd.concat([combined_df_team1, combined_df_team2], ignore_index=True)
```

```python
grouped_df['free_throw_pct_team1'] = grouped_df['total_free_throws_made_team1'] / grouped_df['total_free_throws_attempted_team1']
grouped_df['opponent_free_throw_pct'] = grouped_df['opponent_total_free_throws_made'] / grouped_df['opponent_total_free_throws_attempted']

grouped_df['field_goal_pct_team1'] = grouped_df['total_field_goals_made_team1'] / grouped_df['total_field_goals_attempted_team1']
grouped_df['opponent_field_goal_pct'] = grouped_df['opponent_total_field_goals_made'] / grouped_df['opponent_total_field_goals_attempted']

grouped_df['three_point_field_goal_pct_team1'] = grouped_df['total_three_point_field_goals_made_team1'] / grouped_df['total_three_point_field_goals_attempted_team1']
grouped_df['opponent_three_point_field_goal_pct'] = grouped_df['opponent_total_three_point_field_goals_made'] / grouped_df['opponent_total_three_point_field_goals_attempted']
```

```python
print(np.array(grouped_df.columns))
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

response_var = 'total_wins_team1'
predictor_vars = grouped_df.columns.drop(['team_uid_team1', 'season_team1', response_var])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(grouped_df[predictor_vars])

pca = PCA()
X_pca = pca.fit_transform(X_scaled)

explained_variance = pca.explained_variance_ratio_

# Create a scree plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, 12), explained_variance[:11], marker='o', linestyle='-', label='Individual explained variance')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.title('Scree Plot')
plt.grid(True)
plt.show()

# Use the selected principal components for regression
X_pca_selected = X_pca[:, :7]
y = grouped_df[response_var]

# Fit linear regression model
regression_model = LinearRegression()
regression_model.fit(X_pca_selected, y)

# Predictions and evaluation
y_pred = regression_model.predict(X_pca_selected)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

```
